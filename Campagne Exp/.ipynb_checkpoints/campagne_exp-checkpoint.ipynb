## Jeu de données : MNIST

```python
path = "data"
batch_train_size = 1000
batch_test_size = 1000

train_loader, test_loader = get_dataset(batch_train_size, batch_test_size, path)
```

```python
dim = 28*28
```

## Tests préliminaires

### Perceptron

```python
model = Linear( dim, 10 )
loss = CrossEntropyLoss()
lr = 5e-3
optim = torch.optim.Adam(model.parameters(), lr)

nn = NeuralNetwork(model, loss, optim)
tmp = nn.fit_eval(train_loader, test_loader, n_epochs=100)
plot_CS(*tmp)
```

### Régression logistique

```python
model = Sequential(
    Linear( dim, 10 ),
    torch.nn.Sigmoid()
)
loss = CrossEntropyLoss()
lr = 5e-3
optim = torch.optim.Adam(model.parameters(), lr)

nn = NeuralNetwork(model, loss, optim)
tmp = nn.fit_eval(train_loader, test_loader, n_epochs=100)
plot_CS(*tmp)
```

### First NN - Tanh + Sigmoid

```python
model = Sequential(
    Linear( dim, 15 ),
    torch.nn.Tanh(),
    Linear( 15, 10 ),
    torch.nn.Sigmoid()
)
loss = CrossEntropyLoss()
lr = 5e-3
optim = torch.optim.Adam(model.parameters(), lr)

nn = NeuralNetwork(model, loss, optim)
tmp = nn.fit_eval(train_loader, test_loader, n_epochs=100)
plot_CS(*tmp)
```

### NN - 2 hidden layers

```python
model = Sequential(
    Linear( dim, 15 ),
    torch.nn.Tanh(),
    Linear( 15, 10 ),
    torch.nn.Sigmoid()
)
loss = CrossEntropyLoss()
lr = 1e-4
optim = torch.optim.Adam(model.parameters(), lr)

nn = NeuralNetwork(model, loss, optim)
tmp = nn.fit_eval(train_loader, test_loader, n_epochs=100)
plot_CS(*tmp)
```

### Highway

```python
model = Sequential(
    Linear( dim, 10 ),
    HighwayModel(10),
    HighwayModel(10),
    HighwayModel(10),
)
loss = CrossEntropyLoss()
lr = 1e-4
optim = torch.optim.Adam(model.parameters(), lr)

nn = NeuralNetwork(model, loss, optim)
tmp = nn.fit_eval(train_loader, test_loader, n_epochs=100)
plot_CS(*tmp)
```
